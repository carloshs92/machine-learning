```python
import numpy as np
import matplotlib.pyplot as plt # para hacer plotting
```

```python
# linespace(lim_inferior, lim_superior, numero_elementos)
X = np.linspace(0.4*np.pi, 1000)
print(X[:4])
C = np.cos(X)
print(C[:4])
S = np.sin(X)
```

```python
# PLoteando las funciones sin y cos
plt.figure()
plt.plot(X,C,'r')
plt.plot(X,S,'b')
plt.show()
```

```python
# VISUALIZANDO DATASET DE DIGITOS
from sklearn import datasets
```

```python
digits = datasets.load_digits()
X, Y  = digits.data, digits.target
```

```python
digit = 9
Xdig = X[Y==digit]
np.random.shuffle(Xdig)
```

```python
plt.figure(figsize=(5,5))
for i in range(1, 10):
    plt.subplot(3,3,i)
    plt.axis("off")
    plt.imshow(Xdig[i-1, :].reshape((8,8)), cmap='gray')   
plt.show()    
```

```python
# VISUALIZANDO DATASET DEL IRIS
ds = datasets.load_iris()
X, Y =  ds.data, ds.target
#MOstrando las 5 primeras filas
print(X[0:5,:], Y[0:5,])
```

```python
# Disminuyendo la dimensi√≥n  de la entrada para poder visualizar
X = ds.data[:,np.arange(2)]
```

```python
# MOstrando los datos de forma b√°sico
for output in range(0,3):
    X_iris = X[Y==output,:]
    plt.scatter(X_iris[:,0], X_iris[:,1])
plt.show()    
```
